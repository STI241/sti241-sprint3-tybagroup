```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_distances
import matplotlib.pyplot as plt

df = pd.read_csv('Sprint3.csv', index_col=0).transpose()

def calcular_vecinos(nombre_usuario, k):

    usuario_seleccionado = df.loc[nombre_usuario].values.reshape(1, -1)

    distancias = cosine_distances(usuario_seleccionado, df.values)

    k_indices = distancias.argsort()[0][:k]

    vecinos = df.iloc[k_indices]

    return vecinos, distancias[0, k_indices]

def visualizar_vecinos(vecinos, distancias):
    plt.figure(figsize=(10, 6))
    plt.barh(vecinos.index, distancias, color='skyblue')
    plt.xlabel('Distancia Coseno')
    plt.ylabel('Usuario')
    plt.title('Vecinos más Cercanos')
    plt.gca().invert_yaxis()  
    plt.show()

nombre_usuario = 'Alejandro S'  # Cambiar por el nombre del usuario que se quiera seleccionar
k = 4

vecinos, distancias = calcular_vecinos(nombre_usuario, k)
print("Vecindario de usuarios más cercanos:")
print(vecindario)
print("Distancias:")
print(distancias)

visualizar_vecinos(vecinos, distancias)
```
